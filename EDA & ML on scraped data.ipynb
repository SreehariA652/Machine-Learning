## Supervised learningmodel ##

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.ensemble import RandomForestClassifier #importing Random Fortest Classifier
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
```

```python
df_copy = df.copy()
df_copy
```

```python
from sklearn.preprocessing import LabelEncoder

l = LabelEncoder()

df_copy['Category'] = l.fit_transform(df_copy['Category'])
df_copy['Sub Categories'] = l.fit_transform(df_copy['Sub Categories'])
```

```python
x = df_copy.drop(columns=['Interaction Probability','Product Name'])
x
```

```python
y = df_copy['Interaction Probability']
y
```

## RandomForest ##

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.33,random_state=42)
```

```python
rf=RandomForestClassifier(n_estimators=5,criterion='entropy',random_state=42, max_samples =0.8)
rf.fit(x_train,y_train)
```

```python
pr=rf.predict(x_test)
accuracy_score(y_test,pr)
```

## Hyperparameter tuning ##

```python
error_rate=[]

for i in range(1,40):
    knn_i=KNeighborsClassifier(n_neighbors=i)
    knn_i.fit(x_train,y_train)
    i_pred=knn_i.predict(x_test)
    error_rate.append(np.mean(i_pred != y_test))
```

## Unsupervised learning ##

```python
from sklearn.decomposition import PCA
```

```python
x
```

```python
y
```

```python
pca = PCA(n_components = 2) 
```

```python
x_pca = pca.fit_transform(x)
```

```python
print("Explained Varience Ratio:", pca.explained_variance_ratio_) #to check if the properties(varience) is kept ptoperly even when changed
#explained_variance_ratio_ [0,1] - desired
"""
<0.65 - good
<0.5 - moderate
>3.5 - bad
"""
```

```python
plt.figure(figsize = (10,8))
plt.scatter(x_pca[:,0],x_pca[:,1],c=y)
plt.title('Product data using PCA')
plt.xlabel('Pricipal component 1')
plt.ylabel('Principal component 2')
plt.show()
```

```python
from sklearn.cluster import KMeans
import numpy as np
import pandas as pd
from sklearn.metrics import silhouette_score
```

```python
kmeans = KMeans(n_clusters = 3)
kmeans.fit(x_pca)

labels = kmeans.labels_
silhouette = silhouette_score(x_pca,labels)

print(f"Silhouette Coefficient: {silhouette:.3f}")
```

```python
plt.figure(figsize = (10,8))
plt.scatter(x_pca[:,0],x_pca[:,1],c=labels)
plt.title('Kmeans on product data using PCA')
plt.xlabel('Pricipal component 1')
plt.ylabel('Principal component 2')
plt.show()
```
